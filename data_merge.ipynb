# 파일 병합

```python
path = '/content/gdrive/MyDrive/[광인사] 1차 프로젝트/'

# 파일 병합 함수
def file_merge(path, items_name):
    # 경로에 존재하는 파일들 리스트
    file_list = os.listdir(path+'던파/0801_new/')

    for item_name in items_name:
        # 각 아이템 이름이 들어있는 파일들 호출
        uni = unicodedata.normalize('NFD',f'_{item_name}_')
        item_file = [file for file in file_list if file.find(uni) != -1]

        # 파일들 병합할 빈 DataFrame 생성
        merged_df = pd.DataFrame()

        # 시간별로 추출한 모든 데이터 병합
        for file in item_file:
            df = pd.read_csv(path+'던파/0801_new/' + file, dtype='object')
            merged_df = pd.concat([merged_df, df], sort=False)
        # 중복된 행 제거
        merged_df.drop_duplicates(keep='first', inplace=True)
        merged_df.sort_values(by=['soldDate'], inplace=True)

        # 병합된 파일들 저장
        merged_df.to_csv(path+'코드정리/merged_file/'+item_name+'_merged.csv', index=False, encoding='utf-8-sig')

file_merge(path, items_name)
```
